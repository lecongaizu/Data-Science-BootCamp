```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
import seaborn as sns
sns.set()
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/MyDrive/Data science-BootCamp/Linear Regression/data/real_estate_price_size_year.csv")
```

```python
data.describe()
```

```python
y = data['price']
x1 = data[['size','year']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()  
```
