```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
import seaborn as sns
sns.set()
from google.colab import drive
drive.mount('/content/drive')
```

```python
raw_data = pd.read_csv("/content/drive/MyDrive/Data science-BootCamp/Linear Regression/data/1.03.+Dummies.csv")
```

```python
raw_data
```

```python
data = raw_data.copy();
### Transform data
data['Attendance'] = data['Attendance'].map({'Yes': 1,'No':0})
```

```python
data
```

```python
data.describe()
```

## Regresion

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]

## Using model to calcualte variable
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

## Model


```python
plt.scatter(data['SAT'], y, c = data['Attendance'], cmap='RdYlGn_r')
y_regresion_no = 0.6439 + 0.0014*data['SAT']
y_regresion_yes = 0.8665 + 0.0014*data['SAT']
y_regresion = 0.275 + 0.0017*data['SAT']
fig = plt.plot(data['SAT'], y_regresion_no, lw=2, c='orange', label='regresion line')
fig = plt.plot(data['SAT'], y_regresion_yes, lw=2, c='red', label='regresion line')
fig = plt.plot(data['SAT'], y_regresion, lw=2, c='blue', label='regresion line')
plt.xlabel('SAT',fontsize =20)
plt.ylabel('GPA',fontsize =20)
plt.show()
```
