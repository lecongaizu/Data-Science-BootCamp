### Import libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

### Loading data

```python
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv("/content/drive/My Drive/Data science-BootCamp/Linear Regression/data/1.01.+Simple+linear+regression.csv")
```

```python
x = data['SAT']
y = data['GPA']

x_matrix = x.values.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(x_matrix,y)
```

```python
## R-Squared
reg.score(x_matrix,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x, y)
y_regresion = reg.intercept_ + reg.coef_*x_matrix
fig = plt.plot(x, y_regresion, lw=4, c='orange', label='regresion line')
plt.xlabel('SAT',fontsize =20)
plt.ylabel('GPA')
plt.show()
```
