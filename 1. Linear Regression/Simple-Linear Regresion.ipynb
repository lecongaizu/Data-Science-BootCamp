```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT',fontsize =20)
plt.ylabel('GPA')
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
y_regresion = 0.275 + 0.0017*x1
fig = plt.plot(x1, y_regresion, lw=4, c='orange', label='regresion line')
plt.xlabel('SAT',fontsize =20)
plt.ylabel('GPA')
plt.show()
```
