```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
import seaborn as sns
sns.set()
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/My Drive/Data science-BootCamp/Linear Regression/data/real_estate_price_size.csv")
```

```python
data.describe()
```

```python
y = data['price']
x1 = data['size']
plt.scatter(x1, y)
plt.xlabel('size',fontsize =20)
plt.ylabel('price')
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
y_regresion = 1.019e+05 + 223.1787*x1
fig = plt.plot(x1, y_regresion, lw=4, c='orange', label='regresion line')
plt.xlabel('size',fontsize =20)
plt.ylabel('price')
plt.show()
```
